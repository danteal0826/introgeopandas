```python
import geopandas as gpd
```

```python
localidad=gpd.read_file("/content/barrios")
localidad.to_crs(epsg=9377, inplace=True)
localidad
```

```python
localidad_engativa=localidad[localidad['localidad']=='Engativá']
localidad_engativa
```

```python
from shapely.geometry import Point
casa=Point(-74.124111, 4.726808)
casa
```

```python
casa= gpd.GeoSeries(casa, crs=4326)
casa=casa.to_crs(epsg=9377)
casa
```

```python
mapa=localidad_engativa.plot(figsize=(10,10))
casa.plot(ax=mapa, color="red", marker="*")
```

```python
import matplotlib.pyplot as plt
mapa_2=localidad_engativa.plot(
    column='barriocomu',
    figsize=(15,15),
    edgecolor="black",
    cmap='tab20c',
    legend=True,
    legend_kwds={
        'loc':'upper center',
        'bbox_to_anchor': (0.5, -0.05),
        'ncol':8
    }
)
casa.plot(ax=mapa_2, color="black", marker="*")
plt.title("Mapa localidad Engativá", fontsize=50)
plt.savefig("mapa.pdf", bbox_inches='tight')
```
