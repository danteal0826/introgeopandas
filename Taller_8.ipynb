```python
import geopandas as gpd
```

```python
gdf_localidades = gpd.read_file("/content/drive/MyDrive/Colab Notebooks/localidades.json")
gdf_localidades.to_crs(epsg=9377, inplace=True)
gdf_localidades["Area (Km^2)"] = gdf_localidades.area/1000000
gdf_localidades["Centroide"] = gdf_localidades.centroid
gdf_localidades
```

```python
#Calcular las distancia del centroide de las localidades hasta la univerisdad Distrital
from shapely.geometry import Point
#Point(Longitud, Latitud)
universidad=Point(-74.0657395300252, 4.628174917902711)
universidad
```

```python
#Geodesico (Angulos)
universidad= gpd.GeoSeries(universidad, crs=4326)
universidad
```

```python
#Cartesiano (proyectada)
universidad=universidad.to_crs(epsg=9377)
universidad
```

```python
#Tomar la distancia desde la universidad distrital a cada uno de los puntos del centroide
gdf_localidades["distancia_u"]=[float(universidad.distance(centroid))/1000 for centroid in gdf_localidades.Centroide]
gdf_localidades
```

```python
#Agregar capas
mapa=gdf_localidades.plot(figsize=(20,20))
gdf_localidades["Centroide"].plot(ax=mapa, color="white")
universidad.plot(ax=mapa, color="red", marker="*")
```

```python
import pandas as pd
poblacion=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/osb_demografia-poblacion-localidad.csv", encoding='latin1', delimiter=";")
poblacion.head()
```

```python
#Filtrar solo los datos del 2024
poblacion_2024=poblacion[poblacion["Año"]==2024]
poblacion_2024.head()
```

```python
#Agrupar por localidad
grupos=poblacion_2024.groupby('Localidad')
grupos
```

```python
#Sumar cada localidad
total_poblacion_localidad=grupos["Población"].sum()
print(total_poblacion_localidad)
```

```python
#Convertir la sereie a dataframe
df_poblacion_localidad=total_poblacion_localidad.reset_index()
df_poblacion_localidad
```

```python
#Convertir en mayuscula los nombres de la localidad
df_poblacion_localidad['Localidad']=df_poblacion_localidad['Localidad'].str.upper()
df_poblacion_localidad
```

```python
#Quitarle las tildes al nombre de las localidades
def replace_accents(input_str):
    replacements={
        'Á':'A',
        'É':'E',
        'Í':'I',
        'Ó':'O',
        'Ú':'U'
    }
    for accented_char, unaccented_char in replacements.items():
       input_str=input_str.replace(accented_char, unaccented_char)
    return input_str

df_poblacion_localidad['Localidad']=df_poblacion_localidad['Localidad'].apply(replace_accents)
df_poblacion_localidad

```

```python
filas=df_poblacion_localidad['Localidad']=='LA CANDELARIA'
filas
```

```python
columna='Localidad'
df_poblacion_localidad.loc[filas,columna]="CANDELARIA"
df_poblacion_localidad
```

```python
filas=gdf_localidades['LocNombre']=='LA CANDELARIA'
columna='LocNombre'
gdf_localidades.loc[filas,columna]
```

```python
mergerd_df=gdf_localidades.merge(df_poblacion_localidad, left_on='LocNombre', right_on='Localidad')
mergerd_df.drop('Localidad', axis=1, inplace=True)

mergerd_df
```

#Taller
crear un mapa coropletico de bogota y su poblacion, poner mapa base

```python
import contextily as cx
import matplotlib.pyplot as plt

mapa_3=mergerd_df.plot(
    column='Población',
    figsize=(15, 15),
    edgecolor="black",
    alpha=0.8,
    cmap='Blues',
    legend=True,
    legend_kwds={'label': "Población"}
)

plt.title("Mapa Población Bogota", fontsize=20)
cx.add_basemap(mapa_3, crs=mergerd_df.crs.to_string())
plt.savefig("mapa.pdf", bbox_inches='tight')
```
